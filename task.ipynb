```
from time import sleep
from random import randint
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import urllib.request
import ast
```

```
 
def get(url):
    global ct
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')          
    c=soup.find_all("a","_2UzuFa",href=True)
    for a in c:
        url="https://www.flipkart.com"+a["href"]
        page=requests.get(url)
        soup=BeautifulSoup(page.content,'html.parser')       
        attr_ls=[]
        att=[]
        try:
          try:
             img=soup.find_all(style=re.compile("background-image"))[0]["style"].split("(")[-1].replace(")","")
          except:
            j=soup.find_all(string=re.compile("image"))[0]         
            img=ast.literal_eval(j)[0]["image"]     
          urllib.request.urlretrieve(img,str(ct)+".jpg")
          attr_ls.append("img_path")
          att.append(str(ct)+".jpg")
          ct+=1
          for div in soup.find_all("div","col col-3-12 _2H87wv"):
            if len(div)==0:
              attr_ls.append(" ")
            else:
              for string in div.strings:
                attr_ls.append(string)
          for div2 in soup.find_all("div","col col-9-12 _2vZqPX"):
            for string in div2.strings:
              att.append(string)
          ls=[att]
          cnt=0
          for n,i in enumerate(attr_ls):
            if i==" ":
              attr_ls[n]="drs_atr"+str(cnt)
              cnt+=1
          df=pd.DataFrame(data=ls,columns=attr_ls)
          df_ls.append(df)
          sleep(randint(2,5))
        except:
          pass
```

```
ct=0
df_ls=[]
def get_data(no_page):
  url="https://www.flipkart.com/clothing-and-accessories/pr?sid=clo&sort=price_desc"
  page=requests.get(url)
  soup=BeautifulSoup(page.content,'html.parser')
  b=soup.find_all("a","_2UzuFa",href=True)  
  ls=soup.find("div","_2MImiq").next_element.string.split(" ")
  current=int(ls[1])
  last=int(ls[-1].replace(",",""))
  if (no_page<=last) & (no_page>=1):
    for num in range(1,no_page+1):
      if num==1: 
        get(url)
      else:
        n_url=url+"&page="+str(num)
        get(n_url)
```
